## Building the model

The sigmoid function is defined.

```python
def sigmoid(z):
    d=1+np.e**(-z)
    return 1/d
```

The error function is: -1/N * sigma( y * log(y')+(1-y) * log(1-y')) where y' = w1x1+w2x2+...+wnxn

```python
def errorFunc(X,y,w):
    pred=np.dot(X,w)
    num=(y*np.log(sigmoid(pred))) + ((1-y)*np.log(1-sigmoid(pred)))
    return (-1*sum(num))/len(X)
```

We train the model by updating weights as done in gradient descent. where the final gradient turns out to be x*(yhat-y).
The function retruns weights and losses calculated.

```python
from numpy import random

def train(X,y,epochs=1000,lr=0.01):
    loss=[]
    w=random.rand(X.shape[1])
    N=len(X)
    
    for e in range(epochs):
        pred=sigmoid(np.dot(X,w))
        w = w - ((lr * np.dot(np.transpose(X),pred-y))/N)
        loss.append(errorFunc(X,y,w))
    return [w,loss]
```

This function is to predict.

```python
W,L=train(train_x,train_y,10000,0.05)
```
